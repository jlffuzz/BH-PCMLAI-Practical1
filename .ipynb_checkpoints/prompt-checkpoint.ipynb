### Will a Customer Accept the Coupon?

**Context**

Imagine driving through town and a coupon is delivered to your cell phone for a restaraunt near where you are driving. Would you accept that coupon and take a short detour to the restaraunt? Would you accept the coupon but use it on a sunbsequent trip? Would you ignore the coupon entirely? What if the coupon was for a bar instead of a restaraunt? What about a coffee house? Would you accept a bar coupon with a minor passenger in the car? What about if it was just you and your partner in the car? Would weather impact the rate of acceptance? What about the time of day?

Obviously, proximity to the business is a factor on whether the coupon is delivered to the driver or not, but what are the factors that determine whether a driver accepts the coupon once it is delivered to them? How would you determine whether a driver is likely to accept a coupon?

**Overview**

The goal of this project is to use what you know about visualizations and probability distributions to distinguish between customers who accepted a driving coupon versus those that did not.

**Data**

This data comes to us from the UCI Machine Learning repository and was collected via a survey on Amazon Mechanical Turk. The survey describes different driving scenarios including the destination, current time, weather, passenger, etc., and then ask the person whether he will accept the coupon if he is the driver. Answers that the user will drive there ‘right away’ or ‘later before the coupon expires’ are labeled as ‘Y = 1’ and answers ‘no, I do not want the coupon’ are labeled as ‘Y = 0’.  There are five different types of coupons -- less expensive restaurants (under \\$20), coffee houses, carry out & take away, bar, and more expensive restaurants (\\$20 - \\$50). 

**Deliverables**

Your final product should be a brief report that highlights the differences between customers who did and did not accept the coupons.  To explore the data you will utilize your knowledge of plotting, statistical summaries, and visualization using Python. You will publish your findings in a public facing github repository as your first portfolio piece. 





### Data Description
Keep in mind that these values mentioned below are average values.

The attributes of this data set include:
1. User attributes

    -  Number of times that he/she buys takeaway food: 0, less than 1, 1 to 3, 4 to 8 or greater
    than 8
    -  Number of times that he/she goes to a coffee house: 0, less than 1, 1 to 3, 4 to 8 or
    greater than 8
    -  Number of times that he/she eats at a restaurant with average expense less than \\$20 per
    person: 0, less than 1, 1 to 3, 4 to 8 or greater than 8
    -  Number of times that he/she goes to a bar: 0, less than 1, 1 to 3, 4 to 8 or greater than 8
    

2. Contextual attributes
    - Driving destination: home, work, or no urgent destination
    - Location of user, coupon and destination: we provide a map to show the geographical
    location of the user, destination, and the venue, and we mark the distance between each
    two places with time of driving. The user can see whether the venue is in the same
    direction as the destination.
    - Weather: sunny, rainy, or snowy
    - Temperature: 30F, 55F, or 80F
    - Time: 10AM, 2PM, or 6PM

3. Coupon attributes

3. Decide what to do about your missing data -- drop, replace, other...

While investigating the data, I found a number of columns that should be made into numeric values from objects. These are mostly ranges that can be identified as a single number to help with behavior and relationships. Also a number of entries that were blank that can be at the very least be filled with value of -1 so as to not toss away the other entries within that row and make them easier to disregard when looking at that value. The car column has a lot of empties and a small amount of 5 various inputs. It might be easiest to just disregard that column all together.

```python
# See time values
data['time'].value_counts()
```

```python
# Change time into the 24 hour clock to help with analysis. Use the replace command to change each entry.
data['time'] = data['time'].replace({'7AM': 7, '10AM': 10, '2PM': 14, '6PM': 18, '10PM': 22 })
```

```python
# See the values of expiration
data['expiration'].value_counts()
```

```python
# Replace the two values of expiration with numerical values in hours.
data['expiration'] = data['expiration'].replace({'1d': 24, '2h': 2})
```

```python
# See the values of age
data['age'].value_counts()
```

```python
# Change 50plus to 50 and below21 to 20 to make the column numerical
data['age'] = data['age'].replace({'50plus': 50, 'below21': 20})
# Change the type of the age column to integer using the pandas to_numeric
data['age'] = pd.to_numeric(data['age'])
```

```python
# Take a look at the values in the income column to figure out how to replace ranges with a single number.
data['income'].value_counts()
```

```python
#Remove all the '$' symbols from the values
data['income'] = data['income'].str.replace('$','',regex=False)
```

```python
#Remove words from the two entries and make two entries like the rest of the ranges for next operation
data['income'] = data['income'].replace({'100000 or More': '100000 - 100000', 'Less than 12500': '12500 - 12500'})
```

```python
#A function to split the two numbers of the range and average them to return a single integer
def calc_averageIncome(s):
    nums = s.split(" - ") #delimiter of the range
    num1 = int(nums[0])
    num2 = int(nums[1])
    return round((num1 + num2) / 2)

data['income'] = data['income'].apply(calc_averageIncome) #call function calc_averageIncome to split range
```

```python
#Fill all empty cells in the data frame with '-1~-1'. This will eventually be set to -1 when the function performs below
data = data.fillna('-1~-1')
```

```python
#Replace all worded data for times visited columns to make the field like the others.
data = data.replace({'never': '0~0','less1': '1~1', 'gt8':'8~8'})
```

```python
#After filling in the emptys and changing the worded fields to match together, we can now change the columns data
#to something numerical to help analyze. We will take the ranges and average them. Empty field in the beginning
#will have -1, never = 0, less1 = 1 and gt8 = 8.

#Function to split the two numbers of the string and return an average integer back.
def calc_averageRange(s):
    nums = s.split("~") #delimiter of the range
    num1 = int(nums[0])
    num2 = int(nums[1])
    return round((num1 + num2) / 2)

#Setting the five columns data to numbers.
data['Bar'] = data['Bar'].apply(calc_averageRange)
data['CoffeeHouse'] = data['CoffeeHouse'].apply(calc_averageRange)
data['CarryAway'] = data['CarryAway'].apply(calc_averageRange)
data['RestaurantLessThan20'] = data['RestaurantLessThan20'].apply(calc_averageRange)
data['Restaurant20To50'] = data['Restaurant20To50'].apply(calc_averageRange)
```

1. Create a new `DataFrame` that contains just the bar coupons.


```python
# Get all the rows with coupon equal to Bar and store them in the dataframe barcoups
barcoups = data.query('coupon == "Bar"')
```

2. What proportion of bar coupons were accepted?


```python
#Use the dataframe barcoups and count all the times Y = 1 and then divide the count the total amount 
#of Bar coupons to get the proportion of bar coupons that were accepted
barcoups.query('Y == 1')['Y'].count() / barcoups['Y'].count()
```

### 41% out of the bar coupons offered were accepted.

3. Compare the acceptance rate between those who went to a bar 3 or fewer times a month to those who went more.


```python
# barunder3 is the proportion of those who accepted coupons that were offered for bar visits less than 3
# Also needed to account for the -1 entries added for missing data as the reason we needed to have Bar > 0
barunder3 = barcoups.query('Bar < 3 & Bar > 0').query('Y == 1')['Bar'].count() / barcoups.query('Bar < 3 & Bar > 0')['Bar'].count()

# barover3 is the proportion of those who accepted coupons that were offered for bar visits over 3
barover3 = barcoups.query('Bar > 3').query('Y == 1')['Bar'].count() / barcoups.query('Bar > 3')['Bar'].count()

#Print results
print("The proportion of accepted coupons to offered for under 3 bar visits:",barunder3)
print("The proportion of accepted coupons to offered for over 3 bar visits:",barover3)
```

### Those that accepted the coupon with under 3 visits was 52.7% vs those over 3 visits at 76.9%. Showing that those who attend bars more frequently will more often use the coupon.

4. Compare the acceptance rate between drivers who go to a bar more than once a month and are over the age of 25 to the all others.  Is there a difference?


```python
barover1over25 = barcoups.query('Bar > 1').query('age > 25').query('Y == 1')['Bar'].count() / barcoups.query('Bar > 1').query('age > 25')['Bar'].count()
barunder1n25under = barcoups.query('Bar <= 1 & Bar > 0').query('age <= 25').query('Y == 1')['Bar'].count() / barcoups.query('Bar <= 1 & Bar > 0').query('age <= 25')['Bar'].count()
print("For drivers over 25 and visit the bar more than once a month:",barover1over25)
print("For drivers 25 and under and visit less than once a month:",barunder1n25under)
```

### Though drivers over 25 with more than once a month visit had a larger acceptance of 69.5%. Those 25 and under with less than a visit a month accepted the coupon at 58.3%. One thought I had is that people under a certain age could be interested in savings. It seems like a high amount of acceptance since they visit a bar infrequently.


5. Use the same process to compare the acceptance rate between drivers who go to bars more than once a month and had passengers that were not a kid and had occupations other than farming, fishing, or forestry. 


```python
#Use query command to parse down the data to the scenario then divide the accepted by the total number of the scenario
barOver1NokidsNofff = barcoups.query('Bar > 1').query('passanger != "Kid(s)"').query('occupation != "Farming Fishing & Forestry"').query('Y == 1')['Bar'].count() / barcoups.query('Bar > 1').query('passanger != "Kid(s)"').query('occupation != "Farming Fishing & Forestry"')['Bar'].count()
print('Here is the acceptance rate:',barOver1NokidsNofff)
```

6. Compare the acceptance rates between those drivers who:

- go to bars more than once a month, had passengers that were not a kid, and were not widowed *OR*
- go to bars more than once a month and are under the age of 30 *OR*
- go to cheap restaurants more than 4 times a month and income is less than 50K. 



```python
barOver1NokidsNoWidow = barcoups.query('Bar > 1').query('passanger != "Kid(s)"').query('maritalStatus != "Widowed"').query('Y == 1')['Bar'].count() / barcoups.query('Bar > 1').query('passanger != "Kid(s)"').query('maritalStatus != "Widowed"')['Bar'].count()
barOver1Ageunder30 = barcoups.query('Bar > 1').query('age < 30').query('Y == 1')['Bar'].count() / barcoups.query('Bar > 1').query('age < 30')['Bar'].count()
barCRover4Incomeunder50 = barcoups.query('RestaurantLessThan20 > 4').query('income < 50000').query('Y == 1')['Bar'].count() / barcoups.query('RestaurantLessThan20 > 4').query('income < 50000')['Bar'].count()

print("Acceptance rate for over 1 visit/month, passengers not a kid, and was not widowed:",barOver1NokidsNoWidow)
print("Acceptance rate for over 1 visit/month, and under the age of 30:",barOver1Ageunder30)
print("Acceptance rate for cheap restaurant over 4 visit/month, and income under 50K:",barCRover4Incomeunder50)
```

### We can see that those that go to the bar over once a month tend to accept the coupon for a bar with the first two results over 70%. However, we see a group in the third result that visits cheap restaurants over 4 times a month and the bar coupon acceptance rate falls to under 50%.
